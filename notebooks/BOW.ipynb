# Chosing vector size for dimension reduction

```python
countvec_path = "/content/drive/MyDrive/COMP30027/Assignment/data/ver1/recipe_text_features_countvec"
train_steps_vec = load_npz(os.path.join(countvec_path, "train_steps_vec.npz")).toarray()
train_ingr_vec = load_npz(os.path.join(countvec_path, "train_ingr_vec.npz")).toarray()
train_name_vec = load_npz(os.path.join(countvec_path, "train_name_vec.npz")).toarray()
```

## steps feature

```python
num_features = range(1000, 7000, 1000)
for num_feature in tqdm(num_features):
    pca = PCA(n_components=num_feature)
    pca.fit(train_steps_vec)
    new_data = pca.transform(train_steps_vec)

    X_train, X_test, y_train, y_test = train_test_split(new_data,  train_df["duration_label"], test_size=0.1, random_state=42)
    model = LogisticRegression(random_state=42,  max_iter = 100000)                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                     
    model.fit(X_train, y_train)                                                                                                                                                                                                                                                                                                                                                                              
    y_pred = model.predict(X_test)
    print(classification_report(y_test, y_pred, labels= model.classes_))
```

```python
num_features = range(200, 1200, 200)
for num_feature in tqdm(num_features):
    pca = PCA(n_components=num_feature)
    pca.fit(train_steps_vec)
    new_data = pca.transform(train_steps_vec)

    X_train, X_test, y_train, y_test = train_test_split(new_data,  train_df["duration_label"], test_size=0.1, random_state=42)
    model = LogisticRegression(random_state=42,  max_iter = 100000)                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                     
    model.fit(X_train, y_train)                                                                                                                                                                                                                                                                                                                                                                              
    y_pred = model.predict(X_test)
    print(classification_report(y_test, y_pred, labels= model.classes_))
```
